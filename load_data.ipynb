```python
df_theme = pd.read_csv(os.path.join("/home/kg766/shreyasi_analysis/chatgpt_data/final_analyses/topic_theme_mapping.csv"))
df1['theme'] = df1['Final_Topic_renamed'].map(df_theme[df_theme['country'] == country].set_index('Final_Topic_renamed')['theme'])
```

```python

# rename Final_Topic_renamed to Final_Topic and theme
df1 = df1.rename(columns={
    'convo_id' : 'conversation_id',
    'user_timestamp': 'conversation_timestamp',
    'topic_label_fine': 'Topic_Label_Fine',
    'topic_label_coarse': 'Topic_Label_Coarse',
    'Final_Topic_renamed': 'Clustering_Topic', 
    'theme': 'Clustering_Topic_Theme',
    'label_asking': 'asking_doing_expressing',
    'convo_id' : 'conversation_id',
    'label_work': 'work_coursework_personal',
    'multitasking': 'Multitasking',
    'human_alone': 'Complete_task_alone',
    'user_education': 'User_Education', 
    'AI_education': 'AI_Education'})

df1 = df1.drop(columns=['Final_Topic'])
```

```python
# make rows with values appearing less than 5 times in the Final_Topic column as empty
topic_counts = df1['asking_doing_expressing'].value_counts()
df1.loc[~df1['asking_doing_expressing'].isin(topic_counts[topic_counts >= 5].index), 'asking_doing_expressing'] = None
df1.loc[~df1['work_coursework_personal'].isin(['Work', 'Coursework', 'Personal']), 'work_coursework_personal'] = None
```
